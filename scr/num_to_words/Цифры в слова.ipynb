```python
ad_space = lambda st: ' '+st if st!='' else st

def num_to_words(num):
    s_num = str(num)
    str_out = ''
    for i_gr, gr in enumerate([s_num[::-1][st:st+3] for st in range(0,len(s_num),3)]):
#         print('gr', gr[::-1])
        gr_str = ''
        int_gr = int(gr)
        last_n = ''
        for i, nu in enumerate(gr):
#             В случае если "0" пропускаем но записываем как последенее число
            if nu == '0': 
                last_n = '0'
                continue
#             Обработка крайне левого числа в группе
            if i==0:
                last_n = nu
                gr_str = i_d_dict[int(nu)]
                if i_gr==1: gr_str=gr_str.replace('один','одна').replace('два','две')
                last_n = nu
#             Случаи от 10 до 19
            elif i==1 and nu=='1':
                gr_str = i_d_dict[int(nu+last_n)]
                last_n = ''
#             Остальные числа
            else:
                gr_str = ad_space(gr_str)
                gr_str = f"{i_d_dict[int(nu)*10**i]}{gr_str}"
#         Если ничего не записано пропускаем
        if gr_str=='': continue
#         Случаи до 999 не имеют специального окончания
        if i_gr==0: 
            str_out = gr_str
        else:
            krt=1 # Склонение ранга числа по умолчанию (тысяч и т.д.)
#           Склонение рангов в зависимости от последней цифры
            if i_gr>=1:
                if last_n =='1': krt=0
                elif last_n in ('2', '3', '4'): krt=2                
#             Объединение групп
            str_out = ad_space(str_out)
            str_out = f"{gr_str} {i_gr_dict[i_gr][krt]}{str_out}"
        
    return str_out
```
